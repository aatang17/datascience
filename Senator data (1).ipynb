```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re


url= 'https://www.senate.gov/senators/index.htm'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')




```

```python
Senators = []
State = []
Party  = []

for pages in soup.find_all('tbody'):
    for i in pages.find_all('a', href = True):
        b = i.get_text()
        Senators.append(b)
    
df = pd.DataFrame(Senators, columns = ['Senators'])
        
us_state_abbrev = {
'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA', 'Colorado': 'CO',
'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID',
'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA',
'Maine': 'ME', 'Maryland': 'MD', 'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK',
'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD',
'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA',
'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'} 

my_dict2 = {y:x for x,y in us_state_abbrev.items()}
party = {'R': 'Republican', 'D': 'Democratic', 'I':'Independent'} 
pd.set_option('display.max_rows', 100)       

df.dropna(inplace = True)
new = df['Senators'].str.split(' ', n = 1, expand = True)
df['Last Name'] = new[0]
df['Last Name'] = df['Last Name'].str.strip(',')
df['First Name'] = new [1]
news = df['First Name'].str.split(' ', n= 1, expand = True)
df['First Name'] = news[0]
df['State'] = news[1]
news_ = df['State'].str.split('-', n=1, expand = True)
df['Party'] = news_[0].str.replace('(','')
df['States'] = news_[1].str.replace(')','')
df['First Name'][93] = 'Van Hollen'
df['Last Name'][93] = 'Chris'
#new_final = df['Party'].str.split(' ', n=1 , expand = True)
#df['Political Party'] = new_final[1]   

def split_it(by_party):
    x = re.sub('.*?(?=(R|D|I))', '', by_party)
    return x
df['Party'] = df['Party'].apply(split_it)

df.drop(columns = ['Senators', 'State'], inplace = True)
df['States'] = df['States'].map(my_dict2).fillna(df['States'])
df['Party'] = df['Party'].map(party).fillna(df['Party'])
df.sort_values(by=['States'], inplace=True)
df

from pandas import ExcelWriter
df.to_excel (r'/Users/aarontang/Desktop/Python/senator.xlsx', index =True, header=True)
df
```
